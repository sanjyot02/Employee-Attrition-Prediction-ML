```python
# Importing Modules
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split, GridSearchCV, cross_validate
from sklearn.preprocessing import StandardScaler, MinMaxScaler, LabelEncoder
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, confusion_matrix, roc_curve, auc
from imblearn.under_sampling import RandomUnderSampler
import warnings
warnings.filterwarnings("ignore")
```

```python
# Importing Dataset
df = pd.read_csv(r'C:\Users\sanjy\Downloads\Employee_Attrition_Prediction\WA_Fn-UseC_-HR-Employee-Attrition.csv')
```

```python
df.head()
```

```python
df.info()
```

```python
# Encoding Categorical Values
categorical_cols = ['Attrition', 'BusinessTravel', 'Department', 'Gender', 'JobRole', 'MaritalStatus', 'OverTime', 'EducationField']
encoder = LabelEncoder()
df[categorical_cols] = df[categorical_cols].apply(encoder.fit_transform)
```

```python
df.info()
```

```python
# Define Features and Target
y = df['Attrition']
X = df.drop(['EmployeeCount', 'Attrition', 'EmployeeNumber', 'Over18', 'StandardHours'], axis=1)
```

```python
rus = RandomUnderSampler(random_state=42, replacement=True)
X_, y = rus.fit_resample(X, y)
X = pd.DataFrame(X_, columns=X.columns)
```

```python
# Train-Test-Split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=1)
```

```python
print("After Undersampling:")
print(y.value_counts())
```

```python
# Standardization
standard_scaler = StandardScaler()
X_train_standardized = standard_scaler.fit_transform(X_train)
X_test_standardized = standard_scaler.transform(X_test)
X_standardized = standard_scaler.fit_transform(X)
```

```python
# Normalization
min_max_scaler = MinMaxScaler()
X_train_normalized = min_max_scaler.fit_transform(X_train)
X_test_normalized = min_max_scaler.transform(X_test)
X_normalized = min_max_scaler.fit_transform(X)
```

```python
# Grid Search CV for Hyperparameter Tuning
def tune_hyperparameters(model, X, y):
    param_grid = {
        'C': [0.1, 1, 10, 100],
        'kernel': ['linear', 'rbf']
    }
    grid_search = GridSearchCV(model, param_grid=param_grid)
    grid_search.fit(X, y)
    print("Best Params: ", grid_search.best_params_)
    return grid_search.best_params_
```

```python
print("### Grid Search with Standardized Data ###")
best_parameters_std = tune_hyperparameters(SVC(random_state=0), X_train_standardized, y_train)
```

```python
print("### Grid Search with Normalized Data ###")
best_parameters_norm = tune_hyperparameters(SVC(random_state=0), X_train_normalized, y_train)
```

```python
# Training, Evaluation & ROC Plot
def train_predict_evaluate(model, X_train, y_train, X_test, y_test):
    model.fit(X_train, y_train)
    y_pred = model.predict(X_test)

    print("Accuracy: ", accuracy_score(y_test, y_pred))
    print("Precision: ", precision_score(y_test, y_pred))
    print("Recall: ", recall_score(y_test, y_pred))
    print("F1 Score: ", f1_score(y_test, y_pred))
    print("Confusion Matrix:\n", confusion_matrix(y_test, y_pred))

    fpr, tpr, _ = roc_curve(y_test, y_pred)
    roc_auc = auc(fpr, tpr)

    plt.figure(figsize=(6, 5))
    plt.plot(fpr, tpr, color='blue', label='ROC curve (area = %0.2f)' % roc_auc)
    plt.plot([0, 1], [0, 1], color='grey', linestyle='--')
    plt.xlabel("False Positive Rate")
    plt.ylabel("True Positive Rate")
    plt.title("ROC Curve")
    plt.legend(loc="lower right")
    plt.grid()
    plt.tight_layout()
    plt.show()
```

```python
print("\n### Without Scaling ###\n")
train_predict_evaluate(SVC(random_state=0), X_train, y_train, X_test, y_test)
```

```python
print("\n### With Standardization ###\n")
train_predict_evaluate(SVC(random_state=0, **best_parameters_std), X_train_standardized, y_train, X_test_standardized, y_test)
```

```python
print("\n### With Normalization ###\n")
train_predict_evaluate(SVC(random_state=0, **best_parameters_norm), X_train_normalized, y_train, X_test_normalized, y_test)
```

```python
# K-Fold Cross Validation
def cross_validation(model, X, y):
    scores = cross_validate(model, X, y, cv=5, scoring=('accuracy', 'precision', 'recall', 'f1'))
    metrics = {
        "Accuracy": np.mean(scores['test_accuracy']),
        "Precision": np.mean(scores['test_precision']),
        "Recall": np.mean(scores['test_recall']),
        "F1 Score": np.mean(scores['test_f1'])
    }

    for k, v in metrics.items():
        print(f"{k}: {v:.4f}")
    return list(metrics.values())

metrics = []
```

```python
print("\n### Without Scaling ###\n")
metrics.append(cross_validation(SVC(random_state=0), X, y))
```

```python
print("\n### With Standardization ###\n")
metrics.append(cross_validation(SVC(random_state=0, **best_parameters_std), X_standardized, y))
```

```python
print("\n### With Normalization ###\n")
metrics.append(cross_validation(SVC(random_state=0, **best_parameters_norm), X_normalized, y))
```

```python
# Compare Performances
mdf = pd.DataFrame(metrics, columns=["Accuracy", "Precision", "Recall", "F1 Score"],
                   index=["Without Scaling", "With Standardization", "With Normalization"])

mdf
```

```python
# Plot each metric
fig, axs = plt.subplots(2, 2, figsize=(10, 7))

bar_width = 0.4

for ax, col in zip(axs.ravel(), mdf.columns):
    ax.bar(mdf.index, mdf[col], width=bar_width)
    ax.set_title(col)
    ax.set_ylabel(col)
    ax.set_xticks(range(len(mdf.index)))
    ax.set_xticklabels(mdf.index, rotation=15)
    ax.grid(False)

plt.tight_layout()
plt.show()
```
